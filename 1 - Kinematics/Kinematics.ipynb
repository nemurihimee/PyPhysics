## Key Equations

__Speed__ = Distance/Time

$S = D/t$

__Velocity__ = Displacement/Time

$V = x/t$

__Average Acceleration__ = Velocity/Time

$A_{avg} = V/t$

### 4 Kinematics Equations

__Final Velocity__ = Initial velocity + Acceleration * Time

$V_f = V_i + at$

__Displacement__ = Initial velocity * Time + (1/2) * Acceleration * Time^2

$x = V_it+\frac{1}{2}at^2$

```python
import math

#values
g = 9.81
D = 100   #Distance
t = 10   #Time
x = 100   #Displacement
V = 10   #Velocity
Vi = 5  #Initial velocity
a = 1   #Acceleration
Vf = 0  #Final Velocity

while(D == 0 or t == 0 or x == 0 or V == 0 or Vi == 0 or a == 0 or Vf == 0):
    if D != 0 and t != 0:
        S = D/t     #Speed calculated
        print(S)

    

    if x != 0 and t != 0:
        V = x/t     #Velocity calculated
        print(V)    

    if V != 0 and t != 0:
        Aavg = V/t      #Average acceleration calculated
        print(Aavg)  

    if Vi != 0 and a != 0 and t != 0:
        Vf = Vi+(a*t)       #Final velocity calculated
        print(Vf)

    if Vi != 0 and t != 0 and a != 0:
        x = Vi*t + (1/2)*(a)*(t**2)     #Displacement calculated
        print(x)

    if Vi != 0 and a != 0 and x != 0:
        Vf = math.sqrt(Vi**2 + 2*a*x)   #Final velocity calculated
        print(Vf) 

    if Vf != 0 and Vi != 0 and t != 0:
        x = (1/2)*(Vf+Vi)*t     #Displacement calculated
        print(x)


```
