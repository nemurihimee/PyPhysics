## Key Equations

### Useful Equations

__Speed__ = Distance/Time

$S = \frac{D}{t}$

__Average Velocity__ = Displacement/Time

$\overline{V} = \frac{\Delta x}{t}$

__Acceleration__ = Average Velocity/Time

$a = \frac{\overline{V}}{t}$

### 4 Kinematics Equations

__Velocity at time 't'__ = Initial Velocity + Acceleration * Time

$V_t = V_i + at$

__Displacement__ = (Velocity at time 't' + Initial Velocity) * Time / 2

$\Delta x = (\frac{V_t + V_i}{2})t$

__Displacement__ = Initial Velocity * Time + (1/2) * Acceleration * Time^2

$\Delta x = V_it+\frac{1}{2}at^2$

__Velocity at time 't' squared__ = Initial Velocity squared + 2 * acceleration * Displacement

$V_t^2 = V_i^2 + 2a\Delta x$
